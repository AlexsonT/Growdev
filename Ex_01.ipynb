```python
#4) Exiba o array da questão 2 com todos os números ímpares substituídos por -1 sem alterar o array.

a = np.arange(0,10)

for i in a:
  if i % 2 == 1:
    a[i] = -1
  else:
    a[i] = a[i]  

print(a)
```

```python
#5) Crie 3 matrizes 200 x 200, cada uma com valores aleatórios entre 0 e 255. Junte as 3 matrizes de modo que uma matriz resultante (200 x 200 x 3) seja criada, então exiba como uma imagem utilizando o opencv.
import cv2
from google.colab.patches import cv2_imshow

a = np.random.randint(256, size=(200, 200))
b = np.random.randint(256, size=(200, 200))
c = np.random.randint(256, size=(200, 200))

matriz = np.dstack((a, b, c))

cv2_imshow(matriz)
```

```python
#6) Crie um array bidimensional com numpy (com valores pré-definidos) com M linhas e N colunas e em seguida crie um array com N elementos e verifique se existe uma linha do primeiro array que corresponde ao segundo array criado.
a = np.array([[1, 4, 5],[2, 4, 6], [1, 5, 6]])

b = np.array([1, 5, 6])

existe = False

for i in a:
  if(b == i).all():
    existe = True

print(existe)
```

```python
#7) Crie um array de qualquer dimensão com valores de 1 até 10 e após, altere todos os valores pares para o valor 45.

a = np.arange(1,11)
b = np.where(a % 2 == 1, a, 45)

print(b)
```

```python
#8) Crie um array randômico de tamanho 30 e encontre a média e o desvio padrão.
import statistics
a = np.random.randint(1,30, size=30)

media = statistics.mean(a)
dpad = statistics.pstdev(a)

print('*'*50)
print(f'Os 30 valores gerados são: \n\n{a}')
print('='*50)
print(f'A média da amostra gerada é: {media}')
print(f'O desvio padrão da amostra gerada é: {dpad: .2f}')
print('*'*50)
```

```python
#9) Crie um array randômico de valores inteiros e faça a ordenação desse array.
import statistics
import collections

a = np.random.randint(1, 20, size=18)
ord = np.sort(a)
print(collections.Counter(a))
print(ord)
```

```python
#10)Crie um array randômico de valores inteiros, em seguida encontre a moda.
import numpy as np
import statistics
import collections

a = np.random.randint(1 , 25, size=15)
moda = statistics.mode(a)

print('*'*50)
print(f'Os 15 valores gerados são: \n\n{a}')
print('='*50)
print(f'A moda da amostra gerada é: {moda}')
print(collections.Counter(a))
print('*'*50)
```
